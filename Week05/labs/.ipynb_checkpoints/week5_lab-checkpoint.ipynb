```python
import seaborn as sns
import pandas as pd

diamonds = sns.load_dataset("diamonds")

diamonds.head()
```

#### 1.

- Plot the scatter plot of price and carat. Add clarity into the plot as hue. 

- Do you see any pattern? Is it parallel to your expectation?

```python
plot = sns.scatterplot(x= diamonds.carat, y = diamonds.price, hue = diamonds.clarity, alpha = 0.8)

```

```python

# I find that there are highest number of diamonds that lies in the price range of 10000-18000 with carat between 1-2.5 and more with SI2 clarity
# Also almost all of IF and I1 claritied diamonts are scattered away from the others, which shows that they are less in number 
# I can also see that the price of max. of the diamonds are high in range
# But when I cross checked my obsevation by finding the value range of diamonds with different clarity types, I found that there are more number of diamonds with SI1(brown colour) which does not portray clearly in the plot.
# I can find very less number of SI1 diamonds in the plot which is gainst my expectation 
clarity = diamonds.clarity.value_counts()
clarity
```

#### 2.

- Consider `clarity` and `color` is there any pattern in their heatmap?

```python
plot = sns.displot(diamonds, x="clarity", y="color", cbar = True)
plot.ax.tick_params(labelrotation = 90)
```

```python
# The pattern that I can find from the plot is that all colours of I1 clarity diamonds are less in number
# Also, through the dark shade of the area from the plot we can know that SI1 and VS2 clarity diamonds are more in number
# overall, as the map looks complete without any whitespaces we can know that, there is atleast one diamond that contains all clarity types in all colours.
```
